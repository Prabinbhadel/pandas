```python
csv_1.loc[:, "Date"] = pd.to_datetime(csv_1["Date"])
```

```python
#csv_1['month'] = pd.DatetimeIndex(csv_1['Date']).month
csv_1.loc[:, 'month'] = pd.DatetimeIndex(csv_1['Date']).month
```

```python
csv_1.head()
```

```python
csv_1.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
monthly_sales = csv_1.groupby('month')['Price ($)'].sum()
print(monthly_sales)

# Step 6: Plot the bar graph using matplotlib
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', color='skyblue')
plt.title('Total Sales by Month')
plt.xlabel('month')
plt.ylabel('Price ($)')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines
plt.tight_layout()
plt.show()
```
